```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline
```

```python
def show_gradient_descent(A, b):
#  enter your code here, and remove the following line (30%)
    x = np.array([[-2],[2]])
    alpha = 0.02
    max_iter = 1000
    f = np.linalg.norm(np.dot(A,x)-b)**2
    for k in range(max_iter):
        print(k,('%.04f'% x[0][0], '%.04f'%x[1][0]), '%.08f'%f)
        
        x_prev = x 
        x = x- alpha*2*(A.T@A@x-A.T@b)

        if np.linalg.norm(x-x_prev) <= 10**-8:
            break
        f = np.linalg.norm(np.dot(A,x)-b)**2
```

```python
def draw_gradient_descent(A, b):
    xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
    F1 = A[0][0]*xx + A[0][1]*yy - b[0];
    F2 = A[1][0]*xx + A[1][1]*yy - b[1];
    zz = F1*F1 + F2*F2;
    plt3d = plt.figure().gca(projection='3d')
    plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm)
    #  enter your code here (14%)
    x = np.array([[-2],[2]])
    alpha = 0.02
    max_iter = 1000
   
    for k in range(max_iter):
        ax = plt.gca()
        f  = np.linalg.norm(np.dot(A,x)-b)**2
        ax.scatter(x[0][0], x[1][0],f,c='red')

        x_prev = x 
        x = x- alpha*2*(A.T@A@x-A.T@b) 
        
 
        if np.linalg.norm(x-x_prev) <= 10**-8:
            break
            
    #利用初始值(-3,3)算出x1,x2=-0.9,-1.2 與初始值(-2,2)算出的x1,x2 兩點求出一條線
         
    x = np.linspace(-2,2,100)
    y = 0.5*x-0.75
    
    plt.plot(x, y, color ="blue")
        
    
    plt.show()
```

```python

if __name__ == '__main__':
    b = np.array([[1], [2]])
    A = np.array([[1, -2], [1, -2]])
    show_gradient_descent(A, b)
    draw_gradient_descent(A, b)
```
